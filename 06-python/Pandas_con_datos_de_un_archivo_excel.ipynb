# Pandas

```python
import numpy as np
import pandas as pd
```

## Como leer archivos con panda

```python
# Lectura de un archivo excel
archivo_excel = pd.read_excel('1-2017.xlsx')
```

Pandas transforma los archivos excel en variables de tipo DataFrame. Con ellas podemos realizar distintos tipos de operaciones.

```python
# Como ver lo que hay en un dataframe
archivo_excel
```

```python
# Como transformar un DataFrame a un Array de numpy
arreglo = archivo_excel.to_numpy()
datos = arreglo[1:, :]
```

```python
datos[0:3,1] # Ahora estamos operando con un Array de numpy
```

```python
for fil in range(3):
    print(f'{datos[fil,0]:f} {int(datos[fil,0]):d}')
```

```python
# Seleccionamos datos en el Array de numpy
dia_1 = datos[datos[:,1]==1,0:5]
```
